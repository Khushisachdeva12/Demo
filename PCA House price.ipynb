```python
def fetch_null_count_percent(data):
    nv = data.isnull().sum()
    nv = nv[nv>0]
    nv_res = pd.DataFrame({'Feature':nv.index,'Null_count':nv.values,
                         'Null_percent': (nv.values/data.shape[0])*100})
    nv_res = nv_res.sort_values('Null_percent',ascending = False)
    return nv_res
```

```python
x = df_dum.drop(['Neighborhood','Exterior1st','Exterior2nd','SalePrice'],axis=1)
y = np.log(df_dum['SalePrice'])
print(x.shape)
print(y.shape)
```

### Split data into test and train

```python
from sklearn.model_selection import train_test_split 
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.30,random_state=37)
print(x_train.shape,x_test.shape)
print(y_train.shape,y_test.shape)
```

```python
from sklearn.metrics import *
from sklearn.linear_model import LinearRegression
```

```python
def eval_model(ypred):
    mae = mean_absolute_error(y_test,ypred)
    mse = mean_squared_error(y_test,ypred)
    rmse = np.sqrt(mse)
    print('MAE',mae,'MSE',mse,'RMSE',rmse)
```

#### Linear Reg beore PCA

```python
lr1 = LinearRegression()
lr1.fit(x_train,y_train)
```

```python
ypredlr1 = lr1.predict(x_test)
eval_model(ypredlr1)
```

```python
print('Train_Score',lr1.score(x_train,y_train))
print('Test_Score',lr1.score(x_test,y_test))
```

#### PCA

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
x_train_sc = sc.fit_transform(x_train)
x_test_sc = sc.transform(x_test)
print(x_train_sc.shape)
print(x_test_sc.shape)
```

```python
pca = PCA()

pcomps_train = pca.fit_transform(x_train_sc)
pcomps_test  = pca.transform(x_test_sc)
print(pcomps_train.shape)
print(pcomps_test.shape)
```

```python
eig_val = pca.explained_variance_
eig_vec = pca.components_
```

```python
eig_data_df = pd.DataFrame({'EV':eig_val})
eig_data_df['Var_EV'] = (eig_data_df['EV']/eig_data_df['EV'].sum())*100
eig_data_df['CumSum_var_EV'] = eig_data_df['Var_EV'].cumsum()
eig_data_df.head(10)
```

#### Scree Plot

```python
plt.plot(list(range(156)),eig_data_df['CumSum_var_EV'])
plt.title('Scree Plot')
plt.xlabel('PCs')
plt.ylabel('Var_Eig_Val')
plt.grid()
plt.show()
```
